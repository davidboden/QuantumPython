# Qubit Schrödinger state

The introduction explained that we're going to use the Heisenberg representation to explain what is going on within a quantum circuit. However, we'll break that rule immediately and use the Schrödinger picture to explain what a single qubit looks like. When we're only considering a single qubit, it makes sense to consider its state.

## Classical Bits

A logical bit can take only two values: 0 or 1. In modern silicon chips, a bit is implemented by a tiny region that either lacks or contains roughly 200 electrons. The hardware enforces a clear separation between these two cases—conventionally called 0 (electrons absent) and 1 (electrons present), so that intermediate counts (for example, 10 electrons or 185 electrons in the region) still map unambiguously to logical values 0 or 1. As long as noise does not disturb a large fraction of the charges at once, the system can sample the region, determine the intended value, and repeatedly restore the charge level many times per second to correct any drift. This ability to correct drift and maintain logical states is what makes computers digital.

## Qubits

A qubit has a state that can be represented by a location somewhere on the circumference of a unit circle, which is a circle with radius 1. The horizontal component describes how much $\ket{0}$ is in the state and the vertical component describes how much $\ket{1}$ is in the state. To signify that these are qubit states rather than classical bit states, we add a "ket" around the state $0$ to form $\ket{0}$.

### Qubit states $\ket{+}$ and $\ket{-}$

When the state is equal weights of $\ket{0}$ and $\ket{1}$, the state is $\frac{1}{\sqrt{2}}(\ket{0} + \ket{1})$. If it's not clear why there's a factor of $\frac{1}{\sqrt{2}}$, look at the value of $\sin(\pi/4)$ in radians or $\sin(45°)$, or think about Pythagoras:

<img src="images/qubit-circle-plus.png" alt="qubit circle" width="200"/>

This state is given the label $\ket{+}$ and pronounced "plus". Sometimes it's described as "being in a superposition of $\ket{0}$ and $\ket{1}$", but this is just a trick of language caused by the fact we've chosen $\ket{0}$ as the horizontal basis and $\ket{1}$ as the vertical basis. There's nothing particularly mysterious about a diagonal line.

On the diagram, the state $\ket{-}$, pronounced "minus", is between $\ket{0}$ and $-\ket{1}$. The state is $\frac{1}{\sqrt{2}}(\ket{0} - \ket{1})$.

We can re-draw our qubit by rotating it so that $\ket{+}$ is on the horizontal axis and $\ket{-}$ is the vertical axis. The rotated diagram shows that $\ket{1}$ is the superposition state $\frac{1}{\sqrt{2}}(\ket{+} + \ket{-})$ and $\ket{0}$ is the superposition state $\frac{1}{\sqrt{2}}(\ket{+} − \ket{-})$. The qubit is not usually drawn in this orientation:

## Schrödinger picture of qubit state

Many expositions of quantum mechanics focus on the idea that the qubit state is a vector. This is called the Schrödinger picture. In this picture, the state is represented as a column vector $\begin{bmatrix} \alpha \\ \beta \end{bmatrix}$ where $\alpha$ is the amplitude of $\ket{0}$ and $\beta$ is the amplitude of $\ket{1}$. We've selected the basis vectors of $\ket{0} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ and $\ket{1} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$. Any qubit state can be represented as a linear combination of these two vectors. Because the state lies on the unit circle, $\alpha^2 + \beta^2 = 1$. We will cover the subject of measurement later, but for now let's mention that the probability of obtaining $\ket{0}$ is $\alpha^2$, not $\alpha$ — the amplitude and the probability are different quantities.

```python
from diagrams import buildQubitHTML
from IPython.display import display
from ipywidgets import HTML as HTMLWidget, FloatSlider, VBox
from math import pi

schrodinger_slider = FloatSlider(value=0, min=0, max=2 * pi, step=(2 * pi / 200.0))
schrodinger_output = HTMLWidget(value=buildQubitHTML(schrodinger_slider.value))

def on_slider_change(change):
    schrodinger_output.value = buildQubitHTML(change['new'])

schrodinger_slider.observe(on_slider_change, names='value')
display(VBox([schrodinger_slider, schrodinger_output]))
```

The vectors of the qubit values that we've already covered are:

$\begin{aligned}
  \ket{0} &=                    \begin{bmatrix} 1 \\  0 \end{bmatrix} \\
  \ket{1} &=                    \begin{bmatrix} 0 \\  1 \end{bmatrix} \\
  \ket{+} &= \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\  1 \end{bmatrix} \\
  \ket{-} &= \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ -1 \end{bmatrix}
\end{aligned}$
