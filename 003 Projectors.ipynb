## Projectors

When we measure a qubit, we select two orthogonal (at right angles to each other) vectors and project the qubit's state onto the two vectors to determine which of the chosen measurement basis states the measurement will produce.

The most obvious basis vectors to choose are $\ket{0}$ and $\ket{1}$. This is sometimes given the name Computational basis. If the qubit already happens to be in an exact $\ket{0}$ state or an exact $\ket{1}$ state, then the result of the measurement will always be the corresponding basis state. If the qubit is in the state $\ket{+}$ then we'll see a result of $\ket{0}$ half the time and $\ket{1}$ the other half, at random.

Another basis that you might choose is $\ket{+}, \ket{-}$, which is called the Hadamard basis. 

To calculate the probability of measuring the basis state $\ket{0}$, we *project* the actual state onto $\ket{0}$ and square the length of the result to determine the probability. The reason for the squaring is down to Pythagoras again. If we took just the projected lengths of the state onto the $\ket{0}$ and $\ket{1}$ vectors then they wouldn't sum to 1. The squares sum to 1. Qubits are unit-circles, and Pythagoras is re-stated in the equation of a unit circle: $x^2 + y^2 = 1$

In the simplest case that the qubit happens to already be in either state $\ket{0}$, the probability of finding it in state $\ket{0}$ will be 100% and the probability of finding it in state $\ket{1}$ will be 0%.

Thinking of the qubit as a column vector, we can define a matrix that projects the state onto $\ket{0} = \begin{bmatrix} 1 \\  0 \end{bmatrix}$ as:

$\ket{0}\bra{0} = \begin{bmatrix} 1 \\  0 \end{bmatrix}\begin{bmatrix} 1 & 0 \end{bmatrix} = \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}$

This is the first time we've seen the $\bra{0}$ notation, called "bra" to complete the quip that a bra and a ket together makes a "bracket". The bra is a transposed version of the corresponding ket. In the arrangement above, we have a "ket-bra". Multiplying a vertical matrix and a horizontal matrix of the same dimension is something well-covered in matrix algebra texts; you end up with a square matrix where the value in each position is found by multiplying the corresponding individual inputs from the ket and the bra.

Let's see how this projector acts on example qubit states $\ket{0}$, $\ket{1}$ and $\ket{+}$.

$\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}.\begin{bmatrix} 1 \\  0 \end{bmatrix} = \begin{bmatrix} 1 \\  0 \end{bmatrix}$

$\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}.\begin{bmatrix} 0 \\  1 \end{bmatrix} = \begin{bmatrix} 0 \\  0 \end{bmatrix}$

$\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}.\frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\  1 \end{bmatrix} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\  0 \end{bmatrix}$

These values are amplitudes; square roots of probabilities.

Instead of just multiplying the matrix with the ket, we'll multiply using the arrangement $bra * projector * ket$, which essentially multiplies twice and thereby calculates the square of the result, which is what we need to calculate a probability.

$\begin{bmatrix} 1 & 0 \end{bmatrix}.\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}.\begin{bmatrix} 1 \\ 0 \end{bmatrix} = 1$

^^^ 100% probability of measuring $\ket{0}$ from the starting state of $\ket{0}$

$\begin{bmatrix} 0 & 1 \end{bmatrix}.\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}.\begin{bmatrix} 0 \\ 1 \end{bmatrix} = 0$

^^^ 0% probability of measuring $\ket{0}$ from the starting state of $\ket{1}$

$\frac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \end{bmatrix}.\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}.\frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix} = \frac{1}{2}$

^^^ 50% probability of measuring $\ket{0}$ from the starting state of $\ket{+}$

# Projectors for the Computation basis and Hadamard basis

The four projectors onto the states $\ket{0}$, $\ket{1}$, $\ket{+}$ and $\ket{-}$ are, respectively:

$\ket{0}\bra{0} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}\begin{bmatrix} 1 & 0 \end{bmatrix} = \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}$

$\ket{1}\bra{1} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}\begin{bmatrix} 0 & 1 \end{bmatrix} = \begin{bmatrix} 0 & 0 \\ 0 & 1 \end{bmatrix}$

$\ket{+}\bra{+} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix}\frac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \end{bmatrix} = \frac{1}{2}\begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix}$

$\ket{-}\bra{-} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}\frac{1}{\sqrt{2}}\begin{bmatrix} 1 & -1 \end{bmatrix} = \frac{1}{2}\begin{bmatrix} 1 & -1 \\ -1 & 1 \end{bmatrix}$

# Why does a ket-bra create a projector?

The ket-bra arrangement works because of the way matrix multiplication associates. When a ket-bra acts on a vector, we can group it as the ket acting on the result of the bra acting on the vector:

$$\ket{0}\bra{1}\, v = \ket{0}\!\left(\bra{1}\, v\right)$$

Take a general qubit state $v = \begin{bmatrix} a \\ b \end{bmatrix}$, and consider $\ket{0}\bra{1}$ acting on it:

$$\ket{0}\!\left(\bra{1}\, v\right) = \begin{bmatrix} 1 \\ 0 \end{bmatrix}\!\left(\begin{bmatrix} 0 & 1 \end{bmatrix}\begin{bmatrix} a \\ b \end{bmatrix}\right)$$

The inner part simplifies first. The bra $\bra{1}$ dotted with the vector picks out the coefficient of $\ket{1}$:

$$\bra{1}\, v = \begin{bmatrix} 0 & 1 \end{bmatrix}\begin{bmatrix} a \\ b \end{bmatrix} = b$$

This is just a scalar — it tells us how much of $\ket{1}$ was present in the original vector. Substituting back:

$$\ket{0}\!\left(b\right) = b\begin{bmatrix} 1 \\ 0 \end{bmatrix} = \begin{bmatrix} b \\ 0 \end{bmatrix}$$

The result is $\ket{0}$ scaled by $b$. No matter what vector we started with, the output always points in the $\ket{0}$ direction — and the size of that output depends entirely on how much $\ket{1}$ the original vector contained.

This is the general pattern: a ket-bra $\ket{u}\bra{v}$ applied to any vector first measures "how much of $\ket{v}$ is in that vector" (the bra, producing a scalar), then scales $\ket{u}$ by that amount (the ket). When $u = v$ — as in $\ket{0}\bra{0}$ — the output direction matches what was measured, so the operator is a true projector: it extracts the component along $\ket{0}$ and returns a vector pointing exactly along $\ket{0}$.
