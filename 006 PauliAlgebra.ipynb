# Pauli algebra

After working through unitary transformations in terms of matrix calculations, we can describe the same operations more compactly using Pauli algebra. The three Pauli matrices form a closed algebra under multiplication: any product of Pauli matrices can always be expressed as a scalar multiple of another Pauli matrix or the identity. This means we can manipulate quantum operations symbolically, step by step, without ever expanding out full matrix multiplications.

## The Pauli matrices

The three Pauli matrices are:

$$\sigma_x = \sigma_1 = \begin{bmatrix}0&1\\1&0\end{bmatrix} \qquad \sigma_y = \sigma_2 = \begin{bmatrix}0&-1\\1&0\end{bmatrix} \qquad \sigma_z = \sigma_3 = \begin{bmatrix}1&0\\0&-1\end{bmatrix}$$

> **Note:** The definition of $\sigma_y$ here uses the real form $\begin{bmatrix}0&-1\\1&0\end{bmatrix}$ rather than the standard complex form $\begin{bmatrix}0&-i\\i&0\end{bmatrix}$. The complex version is divided by $i$ to remove complex numbers from the matrix. We do this for simplicity: the algorithms we are explaining here do not involve complex rotations of qubits, so the distinction makes no difference for our purposes.

Each matrix is:
- **Hermitian**: $\sigma_i^\dagger = \sigma_i$ — they are their own conjugate transpose, so they represent physically measurable observables
- **Traceless**: $\text{tr}(\sigma_i) = 0$
- **Involutory**: $\sigma_i^2 = I$ — each matrix is its own inverse

## Multiplication identities

Using the real form of $\sigma_y$, the products contain no imaginary factors:

$$\sigma_x \sigma_y = \sigma_z \qquad \sigma_y \sigma_z = \sigma_x \qquad \sigma_z \sigma_x = -\sigma_y$$

$$\sigma_y \sigma_x = -\sigma_z \qquad \sigma_z \sigma_y = -\sigma_x \qquad \sigma_x \sigma_z = \sigma_y$$

## The Hadamard gate as a Pauli expression

$$H = \frac{1}{\sqrt{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix} \qquad \sigma_x = \begin{bmatrix}0&1\\1&0\end{bmatrix} \qquad \sigma_z = \begin{bmatrix}1&0\\0&-1\end{bmatrix}$$

We can observe that $H = \frac{1}{\sqrt{2}}(\sigma_x + \sigma_z)$.

By observation, $H = \frac{1}{\sqrt{2}}(\sigma_x + \sigma_z)$.

## Transforming $\sigma_z$ under the Hadamard gate

In the previous notebook we verified by matrix multiplication that $H\,\sigma_z\,H = \sigma_x$. We can reach the same result using only Pauli algebra.

The Hadamard is real and symmetric, so $H^\dagger = H$, and the sandwich becomes:

$$H\,\sigma_z\,H = \frac{1}{\sqrt{2}}(\sigma_x + \sigma_z)\;\sigma_z\;\frac{1}{\sqrt{2}}(\sigma_x + \sigma_z)$$

Applying $\sigma_x\sigma_z = \sigma_y$, $\sigma_z^2 = I$, and distributing:

$$= \frac{1}{2}(\sigma_y\sigma_x + \sigma_y\sigma_z + \sigma_x + \sigma_z)$$

Applying $\sigma_y\sigma_x = -\sigma_z$ and $\sigma_y\sigma_z = \sigma_x$:

$$= \frac{1}{2}(-\sigma_z + \sigma_x + \sigma_x + \sigma_z) = \frac{1}{2}(2\sigma_x) = \sigma_x$$

The Z observable becomes the X observable — arrived at purely through Pauli algebra.

## Using SymPy to update observables

So far we have been evaluating Pauli algebra by hand. SymPy can do this automatically. A qubit is represented by its three observables — the X, Y, and Z measurements — held as symbolic Pauli expressions. Applying a gate means evaluating the sandwich $U\,O\,U$ for each observable in turn, which SymPy reduces to a new Pauli expression.

The code below creates a qubit at time zero, whose observables are simply the three Pauli matrices labelled by the qubit name. The Hadamard gate is expressed as the Pauli expression $\frac{1}{\sqrt{2}}(\sigma_x + \sigma_z)$ and passed to `evolve`, which applies the sandwich to each observable and returns a new qubit with updated values.

Later we will write more complex expressions in which a qubit's observables depend on the observables of other qubits — for example, after a two-qubit gate the X observable of one qubit might become a function of the Z observable of another.

```python
from qubit import Qubit
from sympy.physics.paulialgebra import Pauli
from sympy import sqrt
from IPython.display import display

# Express the Hadamard gate as a Pauli expression; label must match the qubit
H_pauli = 1/sqrt(2) * (Pauli(1, label="a") + Pauli(3, label="a"))

a = Qubit.qubit_time_0("a")
display(a)
display(a.evolve(H_pauli))
```
