# Pauli algebra

After working through unitary transformations in terms of matrix calculations, we can describe the same operations more compactly using Pauli algebra. The three Pauli matrices form a closed algebra under multiplication: any product of Pauli matrices can always be expressed as a scalar multiple of another Pauli matrix or the identity. This means we can manipulate quantum operations symbolically, step by step, without ever expanding out full matrix multiplications.

## The Pauli matrices

The three Pauli matrices are:

$$\sigma_x = \sigma_1 = \begin{bmatrix}0&1\\1&0\end{bmatrix} \qquad \sigma_y = \sigma_2 = \begin{bmatrix}0&-1\\1&0\end{bmatrix} \qquad \sigma_z = \sigma_3 = \begin{bmatrix}1&0\\0&-1\end{bmatrix}$$

> **Note:** The definition of $\sigma_y$ here uses the real form $\begin{bmatrix}0&-1\\1&0\end{bmatrix}$ rather than the standard complex form $\begin{bmatrix}0&-i\\i&0\end{bmatrix}$. The complex version is divided by $i$ to remove complex numbers from the matrix. This makes no difference to the algebraic expressions we are manipulating and simplifying, since the Pauli algebra rules capture the symbolic relationships between operators and the imaginary factors cancel in any physically meaningful result.

Each matrix is:
- **Hermitian**: $\sigma_i^\dagger = \sigma_i$ — they are their own conjugate transpose, so they represent physically measurable observables
- **Traceless**: $\text{tr}(\sigma_i) = 0$
- **Involutory**: $\sigma_i^2 = I$ — each matrix is its own inverse

## Multiplication identities

The products follow a cyclic pattern. Going in the order $x \to y \to z \to x$ (forward) gives a factor of $+i$; reversing gives $-i$:

$$\sigma_x \sigma_y = i\sigma_z \qquad \sigma_y \sigma_z = i\sigma_x \qquad \sigma_z \sigma_x = i\sigma_y$$

$$\sigma_y \sigma_x = -i\sigma_z \qquad \sigma_z \sigma_y = -i\sigma_x \qquad \sigma_x \sigma_z = -i\sigma_y$$

Because swapping the order of a product flips the sign, distinct Pauli matrices **anticommute**:

$$\{\sigma_i, \sigma_j\} = \sigma_i \sigma_j + \sigma_j \sigma_i = 0 \quad (i \neq j)$$

Their commutators encode the same cyclic structure:

$$[\sigma_x, \sigma_y] = 2i\sigma_z \qquad [\sigma_y, \sigma_z] = 2i\sigma_x \qquad [\sigma_z, \sigma_x] = 2i\sigma_y$$

```python
from sympy.physics.paulialgebra import Pauli
from IPython.display import display

sigma1 = Pauli(1)  # σ_x
sigma2 = Pauli(2)  # σ_y
sigma3 = Pauli(3)  # σ_z

# Each Pauli matrix squares to the identity
display(sigma1 * sigma1, sigma2 * sigma2, sigma3 * sigma3)

# Cyclic products
display(sigma1 * sigma2, sigma2 * sigma3, sigma3 * sigma1)

# Reversed products (sign flips)
display(sigma2 * sigma1, sigma3 * sigma2, sigma1 * sigma3)
```

## The Hadamard gate as a Pauli expression

The Hadamard gate can be written as a sum of two Pauli matrices:

$$H = \frac{1}{\sqrt{2}}(\sigma_x + \sigma_z)$$

Expanding confirms this:

$$\frac{1}{\sqrt{2}}(\sigma_x + \sigma_z) = \frac{1}{\sqrt{2}}\left(\begin{bmatrix}0&1\\1&0\end{bmatrix} + \begin{bmatrix}1&0\\0&-1\end{bmatrix}\right) = \frac{1}{\sqrt{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix} = H \checkmark$$

## Evolving $\sigma_z$ using Pauli algebra

To evolve the $\sigma_z$ observable under the Hadamard gate we compute $H^\dagger\,\sigma_z\,H$. The Hadamard is a sum of Hermitian Pauli matrices with real coefficients, so $H^\dagger = H$, and the expression becomes $H\,\sigma_z\,H$.

**Step 1** — expand the first product, applying $\sigma_x\sigma_z = -i\sigma_y$ and $\sigma_z\sigma_z = I$:

$$H\,\sigma_z = \frac{1}{\sqrt{2}}(\sigma_x + \sigma_z)\,\sigma_z = \frac{1}{\sqrt{2}}(\sigma_x\sigma_z + \sigma_z\sigma_z) = \frac{1}{\sqrt{2}}(-i\sigma_y + I)$$

**Step 2** — multiply on the right by $H$, applying $\sigma_y\sigma_x = -i\sigma_z$ and $\sigma_y\sigma_z = i\sigma_x$:

$$\frac{1}{\sqrt{2}}(-i\sigma_y + I) \cdot \frac{1}{\sqrt{2}}(\sigma_x + \sigma_z)$$

$$= \frac{1}{2}\bigl(-i\sigma_y\sigma_x - i\sigma_y\sigma_z + \sigma_x + \sigma_z\bigr)$$

$$= \frac{1}{2}\bigl(-i(-i\sigma_z) - i(i\sigma_x) + \sigma_x + \sigma_z\bigr)$$

$$= \frac{1}{2}\bigl(-\sigma_z + \sigma_x + \sigma_x + \sigma_z\bigr) = \frac{1}{2}(2\sigma_x) = \sigma_x$$

The Z observable becomes the X observable — the same result found by matrix multiplication, but arrived at purely through Pauli algebra.

```python
from sympy.physics.paulialgebra import Pauli, evaluate_pauli_product
from sympy import sqrt, expand
from IPython.display import display

sigma1 = Pauli(1)  # σ_x
sigma2 = Pauli(2)  # σ_y
sigma3 = Pauli(3)  # σ_z

H = 1/sqrt(2) * (sigma1 + sigma3)

# Step 1: H * sigma_z
step1 = evaluate_pauli_product(expand(H * sigma3))
display(step1)

# Step 2: (H * sigma_z) * H = sigma_x
result = evaluate_pauli_product(expand(step1 * H))
display(result)
```

```python
from qubit import Qubit
from sympy.physics.paulialgebra import Pauli
from sympy import sqrt

# Express the Hadamard gate as a Pauli expression; label must match the qubit
H_pauli = 1/sqrt(2) * (Pauli(1, label="a") + Pauli(3, label="a"))

a = Qubit.qubit_time_0("a")
display(a)
display(a.evolve(H_pauli))
```
