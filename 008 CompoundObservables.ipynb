# Compound observables

So far each qubit's observables have been Pauli expressions built from its own X, Y and Z. But an observable belonging to one qubit can, after entangling interactions, turn into an expression that also involves *other* qubits' observables.

Suppose qubit a's Z observable has, through some prior entangling gate, become the Pauli expression $x_a\,x_b$ — the product of its own X observable with the X observable of another qubit, $b$. Qubit a's Z observable still depends on its own X observable, $x_a$; it has become entangled with $x_b$ along the way. Measuring qubit a in the computational basis is now a measurement of this **compound observable** $x_a\,x_b$, rather than of $x_a$ alone.

```python
from qubit import Qubit
from sympy.physics.paulialgebra import Pauli
from IPython.display import display

qubit_a = Qubit.qubit_time_0("a").with_xz_observables(
    Pauli(1, label="a"),
    Pauli(1, label="a") * Pauli(1, label="b")
)
display(qubit_a)
```

## What does measuring $x_a x_b$ tell us?

When we measure qubit a, we get a single eigenvalue, $+1$ or $-1$. We do **not** learn the individual eigenvalues of $x_a$ or $x_b$ on their own — only their product. So even though $x_a$ is qubit a's own X observable, entangling it with $x_b$ means a measurement can no longer read it out in isolation.

Each X observable is itself a pair of projectors, one for each of its two possible eigenvalues:

$$\sigma_x^{(a)} = \Pi_+^{(a)} - \Pi_-^{(a)} \qquad \sigma_x^{(b)} = \Pi_+^{(b)} - \Pi_-^{(b)}$$

Multiplying these together as a tensor product shows that the expression comprises four terms:

$$\sigma_x^{(a)} \otimes \sigma_x^{(b)} = \left(\Pi_+^{(a)} - \Pi_-^{(a)}\right) \otimes \left(\Pi_+^{(b)} - \Pi_-^{(b)}\right) = \Pi_+^{(a)}\!\otimes\Pi_+^{(b)} \;+\; \Pi_-^{(a)}\!\otimes\Pi_-^{(b)} \;-\; \Pi_+^{(a)}\!\otimes\Pi_-^{(b)} \;-\; \Pi_-^{(a)}\!\otimes\Pi_+^{(b)}$$

Each of these four terms projects onto one specific combination of $a$ and $b$'s X eigenvalues, and carries the sign of the eigenvalue that combination produces:

* $+\,\Pi_+^{(a)}\!\otimes\Pi_+^{(b)}$ — both $+1$ ($1 \times 1 = 1$)
* $+\,\Pi_-^{(a)}\!\otimes\Pi_-^{(b)}$ — both $-1$ ($-1 \times -1 = 1$)
* $-\,\Pi_+^{(a)}\!\otimes\Pi_-^{(b)}$ — $a$ is $+1$, $b$ is $-1$ ($1 \times -1 = -1$)
* $-\,\Pi_-^{(a)}\!\otimes\Pi_+^{(b)}$ — $a$ is $-1$, $b$ is $+1$ ($-1 \times 1 = -1$)

The two terms with a $+$ sign are exactly the two ways $a$ and $b$ can agree, and the two terms with a $-$ sign are exactly the two ways they can disagree. So:

* Measuring **eigenvalue $+1$** for $x_a x_b$ tells us $a$ and $b$'s X observables came out **the same** as each other — but not which value they shared.
* Measuring **eigenvalue $-1$** for $x_a x_b$ tells us $a$ and $b$'s X observables came out **different** from each other — but not which one was which.

This is the essence of an entangled correlation: the compound observable is sharp (has a definite eigenvalue) even though each individual X observable remains completely un-sharp, a 50/50 toss-up between $+1$ and $-1$.

See [Unknown but equal](https://algassert.com/quantum/2016/01/19/unknown-but-equal.html) for more on this "same or different, but not which" style of correlation.

## Extending to three qubits: $x_a x_b x_c$

The same idea extends to a compound observable built from three X observables, $x_a x_b x_c$.

Each of $a$, $b$ and $c$'s X observable independently comes out $+1$ or $-1$. Multiplying three $\pm 1$ values together gives $+1$ whenever an **even** number of them (zero or two) are $-1$, and gives $-1$ whenever an **odd** number of them (one or three) are $-1$.

So measuring the compound observable $x_a x_b x_c$ tells us the **parity** of how many of the three X observables happened to be $-1$ — nothing more:

* Eigenvalue $+1$: an even number of $a$, $b$, $c$ have X observable $-1$ (none of them, or exactly two of them).
* Eigenvalue $-1$: an odd number of $a$, $b$, $c$ have X observable $-1$ (exactly one of them, or all three).

We still learn nothing about any individual qubit's X value, and nothing about *which* qubits agree or disagree with each other beyond this overall parity — just as with two qubits, only now the correlation is shared across three parties at once.
