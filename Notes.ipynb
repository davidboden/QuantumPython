We assign the z-observable's 1 eigenvalue to be $\ket{0}$ and the -1 eigenvalue to be $\ket{1}$. Although this may seem confusing at first, it at least allows us to think of the "bit value" is an index of -1. With a bit of artistic license, $-1^{\ket{0}} = 1$ eigenvalue and $-1^{\ket{1}} = -1$ eigenvalue.
The z-observables always have expectation value 1 because we've set all qubits to be $\ket{0}$. We can use this to replace Pauli z or 3 matrices in expressions with the identity (matrix) wherever we like.

In Deutsch's paper, the opposite convention is taken. This results in differences to some of the gate behaviour definitions; comments have been made in the Python code to reflect this.
