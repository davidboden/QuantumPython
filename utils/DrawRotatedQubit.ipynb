This code was used to create the image of a qubit rotated so that the $\ket{+}$ state was rightmost in the unit circle. I've just pasted it into the "utils" directory to be out of the way.

```python
from IPython.display import HTML, display
from math import sin, cos, sqrt

def drawQubitPlus():
    phase = 0
    centreX = 80
    centreY = 80
    radius = 50
    display(HTML(f'''
        <svg width="160" height="160">
            <circle cx="{centreX}" cy="{centreY}" r="{radius}" stroke="black" stroke-width="1" fill="white"/>
            <text x="{centreX}" y="{centreY - radius - 15}" dominant-baseline="middle" text-anchor="middle">-|-⟩</text>
            <text x="{centreX}" y="{centreY + radius + 15}" dominant-baseline="middle" text-anchor="middle">|-⟩</text>
            <text x="{centreX + radius + 15}" y="{centreY}" dominant-baseline="middle" text-anchor="middle">|+⟩</text>
            <text x="{centreX - radius - 15}" y="{centreY}" dominant-baseline="middle" text-anchor="middle">-|+⟩</text>
            <text x="{centreX + (1/sqrt(2) * (radius + 15))}" y="{centreY - (1/sqrt(2) * (radius + 15))}" dominant-baseline="middle" text-anchor="middle">|1⟩</text>
            <text x="{centreX - (1/sqrt(2) * (radius + 15))}" y="{centreY + (1/sqrt(2) * (radius + 15))}" dominant-baseline="middle" text-anchor="middle">-|1⟩</text>
            <text x="{centreX + (1/sqrt(2) * (radius + 15))}" y="{centreY + (1/sqrt(2) * (radius + 15))}" dominant-baseline="middle" text-anchor="middle">|0⟩</text>
            <text x="{centreX - (1/sqrt(2) * (radius + 15))}" y="{centreY - (1/sqrt(2) * (radius + 15))}" dominant-baseline="middle" text-anchor="middle">-|0⟩</text>
            <circle cx="{centreX + (radius * cos(phase))}" cy="{centreY - (radius * sin(phase))}" r="5" stroke="black" stroke-width="1" fill="black"/>
        </svg>
    '''))
drawQubitPlus()
```
